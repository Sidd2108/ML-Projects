```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Loading the data
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
credit_card_data.head()
```

```python
credit_card_data.shape
```

```python
credit_card_data.value_counts('Class')# This data is highly unbalanced
```

```python
#separating legit and fraud data
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
print(legit.shape, fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Under Sampling Method to handle unbalanced data
#-> Build a sample dataset containing similar distribution of legit and fraud data
legit_sample = legit.sample(n=492)
```

```python
#Concatenate two dataframe
new_dataset = pd.concat((legit_sample, fraud), axis=0)
new_dataset.to_csv('newcreditcard.csv')
new_dataset
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#Splitting into features and labels
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
#Split into train and test data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
Y_test.value_counts()
```

```python
#Model Training
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train) #Training data
```

```python
X_train_prediction = model.predict(X_train)
```

```python
#Accuracy Score
tr_data_accuracy = accuracy_score(X_train_prediction, Y_train)
tr_data_accuracy #Training data accuracy
```

```python
X_test_prediction = model.predict(X_test)
te_data_accuracy = accuracy_score(X_test_prediction, Y_test)
te_data_accuracy #Test data accuracy
```
